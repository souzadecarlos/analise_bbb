```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from wikitable import wikitable
```

```python
re = requests.get("https://pt.wikipedia.org/wiki/Big_Brother_Brasil_11")
soup = BeautifulSoup(re.text, "html.parser")
```

```python
tables = soup.find_all(class_ = "wikitable")
```

```python
tables[0].find_all("th")[0].text.replace("\n", "")
```

```python
content = []
for i in tables[1].find_all("td"):
    content.append(i.text.replace("\n", ""))
```
